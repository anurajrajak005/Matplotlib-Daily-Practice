```python
df = pd.read_csv("cricket.csv")
df
```

```python
df.columns
```

```python
plt.plot(df['Year'],df['Rohit'],color="red")
plt.plot(df['Year'],df['Virat'],color="Black")
plt.plot(df['Year'],df['Dhoni'],color="green")
plt.xlabel("Year")
plt.ylabel("Score")
plt.show()
```

```python
plt.plot(df['Year'],df['Rohit'],color="red",label="Rohit")
plt.plot(df['Year'],df['Virat'],color="Black",label='Virat')
plt.plot(df['Year'],df['Dhoni'],color="pink",label='Dhoni')
plt.xlabel("Year")
plt.ylabel("Score")
plt.title("Year acc score")
plt.legend()
plt.savefig("Yearscor.jpg")
plt.show()
```

```python
plt.plot(df['Year'],df['Rohit'],color="red",label="Rohit",linestyle="dashed")
plt.plot(df['Year'],df['Virat'],color="Black",label='Virat',linestyle="solid")
plt.plot(df['Year'],df['Dhoni'],color="pink",label='Dhoni',linestyle="dotted")
plt.xlabel("Year")
plt.ylabel("Score")
plt.title("Year acc score")
plt.legend()
plt.savefig("Yearscors.jpg")
plt.show()
```
