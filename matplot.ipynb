```python
Subject = ["C++","Java","SQL","Data Science","cyber"]
marks = [87,67,89,56,77]
e=[0,0,.1,0,.05]
plt.pie(marks,autopct="% 0.1f%%",labels=Subject,radius=1.3,explode=e)
plt.title("Pie Chart")
plt.legend()
# plt.savefig("PIE_CHART")
plt.show()
```

- Stacked graph
- stacked bar plots represents difference groups on the highest of 1 another .
  the peak of the bar depends on the resulting height of the result of the groups.
  It goes from rock bottom to the worth rather then going from zero to value

```python
countries = ["USA","INDIA","CHINA","JAPAN"]
gold = [59,43,50,20]
silver = [120,90,67,82]
bronze = [224,80,170,120]
w=0.4
h=np.add(gold,silver)
a=plt.bar(countries,gold,w,color="red",label="gold",ec="black")
b=plt.bar(countries,silver,w,color="green",bottom=gold,label="silver",ec="black")
c=plt.bar(countries,bronze,w,color="yellow",bottom=h,label="bronze",ec="black")
plt.bar_label(a)
plt.bar_label(b)
plt.bar_label(c)
plt.legend()
plt.show()
```
